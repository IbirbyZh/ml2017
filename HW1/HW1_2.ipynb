```python
from sklearn import datasets, metrics, naive_bayes
from sklearn.model_selection import cross_val_score
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
digits = datasets.load_digits()
breast_cancer = datasets.load_breast_cancer()
sets = [[digits, "digits"], [breast_cancer, "breast_cancer"]]
print digits.data[:5].T
print breast_cancer.data[:5]
```

```python
classifiers = [
                [naive_bayes.BernoulliNB(), "BernoulliNB"],
                [naive_bayes.MultinomialNB(), "MultinomialNB"],
                [naive_bayes.GaussianNB(), "GaussianNB"],
              ]
```

```python
for s in sets:
    for est in classifiers:
        print "%s %s %s"%(est[1], s[1], 
                cross_val_score(est[0], s[0].data, s[0].target).mean(0))
```
